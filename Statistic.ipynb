```python
# Calculate the mean and variance of the "Age" column
mean_age = df["Age"].mean()
variance_age = df["Age"].var()
std_dev_age = np.sqrt(variance_age)  # Standard deviation is the square root of variance

# Plot histogram of Age
sns.histplot(df["Age"], bins=np.arange(25, 51, 2), color="skyblue", kde=True)

# Plot the mean as a red vertical line
plt.axvline(mean_age, color="red", linestyle="--", label=f"Mean Age ({mean_age:.2f})")

# Plot lines for one standard deviation around the mean
plt.axvline(mean_age - std_dev_age, color="orange", linestyle=":", label=f"Mean - 1 Std Dev ({mean_age - std_dev_age:.2f})")
plt.axvline(mean_age + std_dev_age, color="orange", linestyle=":", label=f"Mean + 1 Std Dev ({mean_age + std_dev_age:.2f})")

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution with Mean and Standard Deviation Range")
plt.legend()

# Show the plot
plt.show()
```
