# Explore Data Anaylsis For Imran Khan

```python
dataset["candidate_name"].value_counts()["Imran Khan"]
```

```python
imran_khan_data = dataset[dataset["candidate_name"] == 'Imran Khan']
imran_khan_data
```

```python
win_imran =imran_khan_data[imran_khan_data["outcome"]=="Win"]
sns.barplot(x = "constituency_name" , y = "validated_votes" ,data = win_imran )
plt.show()
```

```python
imran_loss = imran_khan_data[imran_khan_data["outcome"]=="Loss"]
plt.figure(figsize=(25, 14))
sns.barplot(x="constituency_name", y="validated_votes", data=imran_loss, color='red')
plt.xticks(rotation=50, ha='right', fontsize=20)
plt.tight_layout()  
plt.show()
```

```python
grouped_data = imran_khan_data.groupby(['constituency_name', 'outcome'])['validated_votes'].sum().unstack(fill_value=0)

# Plotting
grouped_data.plot(kind='bar', stacked=True, figsize=(30, 15), color=['green', 'red'])
plt.xticks(rotation=50, ha='right', fontsize=25)
plt.xlabel('Constituency Name', fontsize=25)
plt.ylabel('Validated Votes', fontsize=25)
plt.title('Stacked Bar Plot of Validated Votes by Outcome', fontsize=30)
plt.legend(title='Outcome', fontsize=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 6))
ax= sns.countplot(x="outcome", data=imran_khan_data, palette=['green', 'red', 'orange'])
for bars in ax.containers:
    ax.bar_label(bars)

plt.title('Outcome Counts for Imran Khan', fontsize=16)
plt.xlabel('Outcome', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
gb_imran = imran_khan_data.groupby("outcome").agg({"outcome" : "count"})
plt.figure(figsize = (8,8))
plt.pie(gb_imran["outcome"] ,labels = gb_imran.index  , autopct = "%1.2f%%" )
plt.show()
```

# In 1993 Election data insight for the punjab election

```python
 data_1993=pml_nawaz[pml_nawaz["election_date"] == "10/6/93"]
punjab_data=data_1993[data_1993["province"]=="Punjab"]
```

```python
grouped_data = punjab_data.groupby(['candidate_name', 'outcome'])['candidate_share'].sum().unstack(fill_value=0)

plt.figure(figsize=(25, 15))
grouped_data.plot(kind='bar', stacked=True, figsize=(25, 15), color=sns.color_palette("viridis", len(grouped_data.columns)))

plt.xticks(rotation=80, ha="right", fontsize=20)
plt.yticks(fontsize=14)
plt.xlabel("Candidate Name", fontsize=10) 
plt.ylabel("candidate_share", fontsize=16)  
plt.title("candidate_share by Candidate in 1993", fontsize=20)
plt.show()
```

## PML(N) for Sindh

```python
indh_data = data_1993[data_1993["province"] == "Sindh"]

grouped_data = sindh_data.groupby(['candidate_name', 'outcome'])['candidate_share'].sum().unstack(fill_value=0)

plt.figure(figsize=(25, 15))
grouped_data.plot(kind='bar', stacked=True, figsize=(25, 15))

plt.xticks(rotation=80, ha="right", fontsize=20)
plt.yticks(fontsize=14)
plt.xlabel("Candidate Name", fontsize=10)
plt.ylabel("candidate_share", fontsize=16)
plt.title("candidate_share by Candidate in Sindh (1993)", fontsize=20)
plt.legend(title='Outcome', fontsize=14)
plt.tight_layout()
plt.show()
```

## PML(N) Election for Khyber Paktunkhwa (NWFP)

```python
pk_data = data_1993[data_1993["province"] == "Khyber Paktunkhwa (NWFP)"]

grouped_data = kpk_data.groupby(['candidate_name', 'outcome'])['validated_votes'].sum().unstack(fill_value=0)

plt.figure(figsize=(25, 15))
grouped_data.plot(kind='bar', stacked=True, figsize=(25, 15))

plt.xticks(rotation=80, ha="right", fontsize=20)
plt.yticks(fontsize=14)
plt.xlabel("Candidate Name", fontsize=10)
plt.ylabel("Validated Votes", fontsize=16)
plt.title("Validated Votes by Candidate in Khyber Paktunkhwa (1993)", fontsize=20)
plt.legend(title='Outcome', fontsize=14)
plt.tight_layout()
plt.show()
```
